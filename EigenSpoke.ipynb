```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
G = nx.newman_watts_strogatz_graph(250,4,0.05)
```

```python
G = nx.Graph(G)
A = nx.to_numpy_matrix(G)
```

```python
A
```

```python
u, s, v = np.linalg.svd(A)
```

```python
np.reshape(np.array(u[0]),-1)
```

```python
f, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3,  sharex='col', sharey='row', figsize=(8,8), dpi=80)
ax1.scatter(np.reshape(np.array(u[0]), -1), np.reshape(np.array(u[1]),-1))
ax2.scatter(np.reshape(np.array(u[1]), -1), np.reshape(np.array(u[2]), -1))
ax3.scatter(np.reshape(np.array(u[2]),-1), np.reshape(np.array(u[3]),-1))
ax4.scatter(u[3], u[4])
ax5.scatter(u[4], u[5])
ax6.scatter(u[5], u[6])
ax7.scatter(u[6], u[7])
ax8.scatter(u[7], u[8])
ax9.scatter(u[8], u[9])

plt.show()
```
